# 1)

```python
# Read the text file
with open('Corpus.txt', 'r', encoding='utf-8') as file:
    corpus_text = file.read()

# Tokenize the text into words
tokens = word_tokenize(corpus_text)

# Number of tokens (total words)
num_tokens = len(tokens)

# Number of unique words
num_unique_words = len(set(tokens))

# Count the occurrences of each word
word_counts = Counter(tokens)

print("A) Number of words in the text (tokens):", num_tokens)
print("B) Number of unique words used:", num_unique_words)
print("C) Number of repetitions of each word:")
for word, count in word_counts.items():
    print(f"{word}: {count}")
```

```python
def filter_tokens(tokens):
    
    filtered_tokens = []
    for word in tokens:
        # Check if the word contains parentheses and the character "س"
        if '(' in word and ')' in word and 'س' in word:
            filtered_tokens.append(word)
        # Skip tokens that contain parentheses but don't contain "س"
        elif '(' in word or ')' in word:
            continue
        # Add other tokens to the filtered list
        else:
            filtered_tokens.append(word)
    return filtered_tokens

def fix_punctuation(text):
    
    # Remove extra spaces before punctuation marks
    punctuation_marks = ['.', '،', '؛', ':', '!', '?', '(',')']
    for mark in punctuation_marks:
        text = text.replace(' ' + mark, mark)
        
    # Remove extra spaces after punctuation marks
    for mark in punctuation_marks:
        # Exclude ')' from removing extra spaces after it
        if mark != ')':
            text = text.replace(mark + ' ', mark)

    # Add space after punctuation marks except for '(' and '.'
    for i, mark in enumerate(punctuation_marks):
        if mark != '(' and mark != '.':
            text = text.replace(mark, mark + ' ')
        elif mark == '.' and i > 0 and punctuation_marks[i - 1] != '.':
            text = text.replace(mark, mark + ' ')

    
    # Tokenize the text
    tokens = text.split()
    
    # Remove []
    tokens = [word for word in tokens if not (word.startswith('[') or word.endswith(']'))]

    # Filter tokens
    filtered_tokens = filter_tokens(tokens)
    
    # Join filtered tokens back into text
    text = ' '.join(filtered_tokens)
    
    return text


# Function to preprocess the text
def preprocess_text(text):
    
    # You can add your preprocessing steps here if needed
    return fix_punctuation(text)

# Preprocess the text
processed_text = preprocess_text(corpus_text)

print("Original text:", corpus_text)
print("Processed text:", processed_text)
```
